```python
import sklearn as sk
import pandas as pd
from sklearn.model_selection import train_test_split #We need this to split the data
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Speed Dating Data.csv',encoding='latin-1')
df.head()
```

```python
# Which dataset and variables you selected and why : I selected go_out, age, field_cd, race, income, goal, career_c
#I prefered using decision tree rather random forest in this assignment. 
```

```python
def correctIncome(x):
    x = x.replace(",","")
    return x
```

```python
newdf = df.loc[:, ["go_out", "age", "field_cd", "race", "income", "goal", "career_c", "dec_o"]]
newdf = newdf.dropna() # to get rid of NA in career_c 
#newdf.reset_index(inplace = True) # note to myself to reset the index from 0 to the last row, same as newdf = newdf.reset_index()
newdf = newdf.reset_index()
newdf["income"] = newdf["income"].apply(correctIncome) #I had to delete the comma because an error happened in my code
```

```python
X= newdf.loc[:, "go_out":"career_c"]
y= newdf.loc[:, ["dec_o"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
```

```python
dtree = DecisionTreeClassifier() #create the tree
dtree = dtree.fit(X_train, y_train) #fit the tree with the train data
```
